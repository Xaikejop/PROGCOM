```python
import random

class Pokemon:
    def __init__(self, nombre, tipo, hp):
        self.nombre = nombre
        self.tipo = tipo
        self.hp = hp
        self.max_hp = hp

    def atacar(self, otro_pokemon):
        daño = random.randint(10, 20)  # Daño aleatorio entre 10 y 20
        otro_pokemon.recibir_daño(daño)
        print(f"{self.nombre} ataca a {otro_pokemon.nombre} y le causa {daño} de daño.")

    def recibir_daño(self, cantidad):
        self.hp -= cantidad
        if self.hp < 0:
            self.hp = 0

    def esta_debilitado(self):
        return self.hp == 0

    def curar(self):
        self.hp = self.max_hp
        print(f"{self.nombre} ha sido curado completamente.")


class PokemonAgua(Pokemon):
    def __init__(self, nombre, hp):
        super().__init__(nombre, "Agua", hp)

    def ataque_especial(self, otro_pokemon):
        daño = random.randint(15, 25)
        otro_pokemon.recibir_daño(daño)
        print(f"{self.nombre} usa Pistola Agua y le causa {daño} de daño a {otro_pokemon.nombre}.")


class PokemonFuego(Pokemon):
    def __init__(self, nombre, hp):
        super().__init__(nombre, "Fuego", hp)

    def ataque_especial(self, otro_pokemon):
        daño = random.randint(15, 25)
        otro_pokemon.recibir_daño(daño)
        print(f"{self.nombre} usa Ascuas y le causa {daño} de daño a {otro_pokemon.nombre}.")


class Entrenador:
    def __init__(self, nombre):
        self.nombre = nombre
        self.pokemons = []

    def capturar_pokemon(self, pokemon):
        self.pokemons.append(pokemon)
        print(f"{self.nombre} ha capturado a {pokemon.nombre}!")

    def seleccionar_pokemon(self, indice):
        if 0 <= indice < len(self.pokemons):
            return self.pokemons[indice]
        else:
            print("Índice de Pokémon inválido.")
            return None

    def liberar_pokemon(self, pokemon):
        if pokemon in self.pokemons:
            self.pokemons.remove(pokemon)
            print(f"{self.nombre} ha liberado a {pokemon.nombre}.")
        else:
            print(f"{pokemon.nombre} no está en tu equipo.")



```
